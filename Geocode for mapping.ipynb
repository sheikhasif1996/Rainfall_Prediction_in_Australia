```python
#impoting necessary libraries
import pandas as pd
import geopandas as gpd
from geopandas.tools import geocode
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="Sheikh Asif")
import folium
from shapely.geometry import Point
```

```python
#import data
raw_data = pd.read_csv('weatherAUS.csv')
```

```python
raw_data.head()
```

```python
Location = raw_data['Location'].unique()
```

```python
Location
```

```python
new_df = raw_data[['Location','Rainfall']]
```

```python
rainfall_dict={}
for loc in Location:
    rainfall_dict[loc]=new_df.Rainfall[new_df.Location==loc].mean()
    
```

```python
rainfall_dict
```

```python
mean_rainfall_df= pd.DataFrame(list(rainfall_dict.items()), columns=['Location', 'Mean Rainfall'])
```

```python
mean_rainfall_df
```

```python
geolocator = Nominatim(user_agent="Sheikh Asif")
for index, row in mean_rainfall_df.iterrows():
    try: 
        print(row['Location'])
        info = geocode(str(row['Location']))
        Location = info['geometry'].loc[0]
        if Location.is_empty == True:
            Location = Point(0, 0)
        else:
            mean_rainfall_df.loc[int(index), 'Address'] = info['address'].loc[0]
            mean_rainfall_df.loc[int(index), 'Lon'] = Location.x
            mean_rainfall_df.loc[int(index), 'Lat'] = Location.y        
    except TypeError:
        print ('Geocoding information for '+ row['Location']+ " is not found!")
        print ('')
```
